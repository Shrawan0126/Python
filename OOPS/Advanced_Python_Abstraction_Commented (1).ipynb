## Abstract Base Classes (ABC)

```python
# ABSTRACT BASE CLASS (ABC) IN PYTHON

# ABC = Abstract Base Class
# It is used to create a blueprint (template) for other classes.
#
# An abstract class:
# - Can contain abstract methods.
# - Cannot be instantiated (you cannot create its object).
# - Forces child classes to implement certain methods.

# Importing ABC and abstractmethod

from abc import ABC, abstractmethod

# ABC stands for Abstract Base Class.
# By inheriting from ABC, this class becomes an abstract class.

# abstractmethod
# It is a decorator.
# Any method marked with @abstractmethod MUST be implemented by every child class.


# Creating an Abstract Class

class Shape(ABC):
    # Abstract Method: area()
    # Every child class MUST define its own area() method.
    # We don't know the formula here because different shapes have different formulas.

    @abstractmethod
    def area(self):
        pass      # Placeholder (does nothing)

    
    # Abstract Method: perimeter()
    # Every child class MUST define its own perimeter() method.

    @abstractmethod
    def perimeter(self):
        pass


# This is NOT allowed

# shape = Shape()

# If you uncomment the above line, Python gives:

# TypeError:
# Can't instantiate abstract class Shape with abstract methods area, perimeter

# Why?
# Because Shape is incomplete.
# It only acts as a blueprint.



# Child Class 1 : Rectangle

class Rectangle(Shape):

    # Constructor
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # Implementing area()
    # Since Shape requires it, we MUST define it.
    def area(self):
        return self.width * self.height

    # Implementing perimeter()
    def perimeter(self):
        return 2 * (self.width + self.height)



# Child Class 2 : Circle

class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    # Circle's area formula
    def area(self):
        return 3.14 * self.radius ** 2

    # Circle's perimeter formula
    def perimeter(self):
        return 2 * 3.14 * self.radius


# Creating Objects

# Since Rectangle implements all abstract methods,
# Python allows us to create its object.

rectangle = Rectangle(10, 5)

# Since Circle also implements all abstract methods,
# Python allows this too.

circle = Circle(7)


# Calling Methods

print("Rectangle")
print("Area:", rectangle.area())
print("Perimeter:", rectangle.perimeter())

print()

print("Circle")
print("Area:", circle.area())
print("Perimeter:", circle.perimeter())


# Example of an Incorrect Child Class

# Uncomment the code below to see the error.

"""
class Triangle(Shape):

    def area(self):
        return 100

    # Forgot to implement perimeter()


triangle = Triangle()

# Output:
# TypeError:
# Can't instantiate abstract class Triangle
# with abstract method perimeter
"""


# Why Use Abstract Classes?
# Imagine you are making a drawing application.

# Shapes:
# Rectangle
# Circle
# Triangle
# Square
# Pentagon

# Every shape should have:
#
# area()
# perimeter()

# Instead of reminding every programmer,
# Python enforces the rule using abstract classes.

# If a programmer forgets to write one of these methods,
# Python immediately shows an error.

# This makes code:
#
# - Safer
# - Easier to maintain
# - More organized
# - More consistent


# Real-Life Analogy

# Think of Shape as a blueprint for houses.

# Blueprint says:
#
# Every house must have:
# - Door
# - Roof
# - Windows
#
# But it doesn't specify their design.

# Different builders make different houses.

# Likewise,

# Shape says:
#
# Every shape must have:
# - area()
# - perimeter()
#
# But Rectangle, Circle, Triangle, etc.
# each provide their own implementation.


# Summary

# ABC
# Makes a class abstract.

# @abstractmethod
# Forces child classes to implement the method.

# Abstract Class
# Cannot create its object.

# Child Class
# Must implement every abstract method.

# Benefit
# Ensures every child class follows the same structure.
```

## Implementing Abstract Methods

```python

# Rectangle is a concrete class because it implements
# ALL abstract methods defined in Shape.

class Rectangle(Shape):

    def __init__(self, length, width):
        # Store instance attributes
        self.length = length
        self.width = width

    def area(self):
        # Formula:
        # Area = Length × Width
        return self.length * self.width

    def perimeter(self):
        # Formula:
        # Perimeter = 2 × (Length + Width)
        return 2 * (self.length + self.width)


# Create an object
rect = Rectangle(10, 5)

print("Area =", rect.area())
print("Perimeter =", rect.perimeter())
```

## Abstract Method with Default Implementation

```python

from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        # Abstract methods may still contain code.
        # Child classes can reuse this logic.
        print("Running basic startup sequence...")


class Car(Vehicle):

    def start(self):
        # Reuse parent implementation.
        super().start()

        # Add custom functionality.
        print("Starting fuel injection system")
        print("Car Started Successfully")


car = Car()
car.start()
```

## Abstract Properties

```python
class Employee(ABC):

    @property
    @abstractmethod
    def salary(self):
        '''
        Every employee must expose salary.
        '''
        pass


class Developer(Employee):

    @property
    def salary(self):
        # Computed salary could come from
        # a database, API, payroll service, etc.
        return 120000


dev = Developer()

print("Developer Salary =", dev.salary)
```

## Runtime Polymorphism Using Abstraction

```python

class PaymentGateway(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


class CreditCard(PaymentGateway):

    def pay(self, amount):

        # Credit card specific implementation
        print(f"Paid ₹{amount} using Credit Card")


class UPI(PaymentGateway):

    def pay(self, amount):

        # UPI specific implementation
        print(f"Paid ₹{amount} using UPI")


def process_payment(payment_method, amount):

    # Notice:
    # We don't care whether the object is CreditCard,
    # UPI, PayPal, etc.
    #
    # We only care that it follows the PaymentGateway
    # contract.
    payment_method.pay(amount)


process_payment(CreditCard(), 500)
process_payment(UPI(), 1000)
```

## Template Method Design Pattern

```python

class DataPipeline(ABC):

    # Template Method
    #
    # Defines the overall algorithm.
    # Subclasses customize individual steps.
    def run(self):

        self.extract()
        self.transform()
        self.load()

    @abstractmethod
    def extract(self):
        pass

    @abstractmethod
    def transform(self):
        pass

    @abstractmethod
    def load(self):
        pass


class CSVPipeline(DataPipeline):

    def extract(self):

        # Read data from source
        print("Reading CSV file")

    def transform(self):

        # Clean and prepare data
        print("Transforming data")

    def load(self):

        # Save data to destination
        print("Loading into database")


pipeline = CSVPipeline()
pipeline.run()
```

## Protocols (Structural Typing)

```python

from typing import Protocol


# Protocol checks behavior rather than inheritance.
class Flyable(Protocol):

    def fly(self) -> str:
        ...


class Bird:

    def fly(self):

        return "Bird is flying"


class Drone:

    def fly(self):

        return "Drone is flying"


def launch(obj: Flyable):

    # Any object having fly() can be passed.
    print(obj.fly())


launch(Bird())
launch(Drone())
```

## Plugin Architecture Example

```python

class Plugin(ABC):

    @abstractmethod
    def execute(self):
        pass


class EmailPlugin(Plugin):

    def execute(self):

        # Email specific behavior
        print("Sending Email Notification")


class SMSPlugin(Plugin):

    def execute(self):

        # SMS specific behavior
        print("Sending SMS Notification")


class Application:

    def __init__(self):

        # Store all registered plugins
        self.plugins = []

    def register(self, plugin):

        # Add plugin to system
        self.plugins.append(plugin)

    def run(self):

        # Execute every plugin
        for plugin in self.plugins:
            plugin.execute()


app = Application()

app.register(EmailPlugin())
app.register(SMSPlugin())

app.run()
```

## Key Interview Notes

### Abstraction
Hides implementation details and exposes only essential functionality.

### Encapsulation vs Abstraction

Encapsulation:
- Hides data.

Abstraction:
- Hides complexity.

### ABC vs Protocol

ABC:
- Requires inheritance.
- Runtime enforcement.

Protocol:
- No inheritance required.
- Structural typing.
