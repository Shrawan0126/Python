Encapsulation is the concept of wraping data(variable) and methods(function) together as a simgle unit.It restricts direct access to some the objs components, which is a means of preventing accidental interference and misuse of the data

```python
## Public, protected, private
class Person:
    def __init__(self,name,age):
        self.name = name    ## public variables
        self.age = age

def get_name(person):
    return person.name

person = Person("Krish", 22)
get_name(person)
```

```python
class Person:
    def __init__(self,name,age,gender):
        self._name = name    ## private variables
        self._age = age       ## Protected variable
        self.gender = gender

class Employee(Person):
    def __init__(self,name,age,gender):
        super().__init__(name,age,gender)

employee = Employee("Krish", 22,"Male")
print(employee._name)
```
