```python
# Define a custom exception
class dobException(Exception):
    pass

year = int(input("Enter the dob: "))
age = 2026 - year

try:
    if 20 <= age <= 30:
        print("The age is valid so you can apply for the exam")
    else:
        raise dobException
except dobException:
    print("Sorry, your age not in range")
```
