Logging with Multiple Loggers

You can create Multiple loggers for different parts of your application

```python
import logging  # Import Python's logging module

# Configure the root logging system
logging.basicConfig(
    filename = 'app2.log',
    filemode = 'w',
    level=logging.DEBUG,  # Root logger handles DEBUG and above
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',  # Log format
    datefmt='%Y-%m-%d %H:%M:%S'  # Date/time format
)
```

```python
## log message with different loggers
# Create a logger for module1
logger1 = logging.getLogger("module1")  # Named logger "module1"
logger1.setLevel(logging.DEBUG)         # Allows DEBUG and all higher levels

# Create a logger for module2
logger2 = logging.getLogger("module2")  # Named logger "module2"
logger2.setLevel(logging.WARNING)       # Allows only WARNING and above

# Log messages using logger1
logger1.debug("This is a DEBUG message from module1")   # Will be shown
logger1.info("This is an INFO message from module1")    # Will be shown
logger1.warning("This is a WARNING message from module1")  # Will be shown

# Log messages using logger2
logger2.debug("This DEBUG message from module2 will NOT be shown")  # Blocked
logger2.info("This INFO message from module2 will NOT be shown")    # Blocked
logger2.warning("This is a WARNING message from module2")  # Will be shown
logger2.error("This is an ERROR message from module2")     # Will be shown
```
