Logging in Python is a way to keep a record of what your program is doing while it runs. It helps you track important events, messages, and errors so you can understand how your program behaves and fix problems if something goes wrong. Instead of just printing messages on the screen, logging allows you to organize and save these messages, making it easier to monitor and debug your code, especially in larger projects like AI and machine learning.


Why use logging?
- Track program execution
- Debug issues
- Monitor errors in production
- Save logs to files instead of just printing to screen
- Control how much information you want to see

Log Levels
Python logging module has several log levels indicating the severity of events. The default levels are : 

- DEBUG: Detailed information, typically of interest only when diagnosing problems.
- INFO: Confirmation that things are working as expected.
- WARNING: An indication that something unexpected happened or indicative of some problem in the near future (e.g., 'disk space low'). The software is still working as expected.
- ERROR: Due to a more serious problem, the software has not been able to perform some function.
- CRITICAL: A very serious error, indicating that the program itself may be unable to continue running.



```python
import logging
```

```python
import logging  # Import the logging module

# Configure logging settings
# level=logging.DEBUG means all messages from DEBUG and above will be shown
logging.basicConfig(level=logging.DEBUG)

# Log a debug message (lowest level, used for detailed debugging)
logging.debug("This is a debug message")

# Log an informational message (general program updates)
logging.info("This is an info message")

# Log a warning message (something unexpected but not an error)
logging.warning("This is a warning message")

# Log an error message (a serious issue occurred)
logging.error("This is an error message")
```

```python
import logging  # Import logging module

# Configure logging settings : customizing the logging system
logging.basicConfig(
    filename='app.log',  # Log file name where messages will be stored
    filemode='w',        # 'w' = overwrite file each run
    level=logging.DEBUG, # Capture all levels (DEBUG and above)
    
    # Define log message format
        # %(asctime)s → Timestamp
        # %(name)s → Logger name (usually "root")
        # %(levelname)s → DEBUG, INFO, etc.
        # %(message)s → Your actual message
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    
    # Define how date/time should appear
    datefmt='%Y-%m-%d %H:%M:%S'
)

# Log messages of different severity levels

logging.debug("This is a debug message")       # Detailed debugging info
logging.info("This is an info message")        # General program info
logging.warning("This is a warning message")   # Something unexpected
logging.error("This is an error message")      # A serious issue occurred
logging.critical("This is a critical message") # Very serious error
```
