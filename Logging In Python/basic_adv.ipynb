1. What is Logging (Basic Idea)

    Logging = recording what your program is doing

    Instead of:

print("Something happened")

    We use:

logging.info("Something happened")

Difference:
print() → temporary output
logging → structured, controllable, storable

2. Why Logging is Needed
    - Without logging:
        - Hard to debug
        - No history
        - No severity levels
    - With logging:
        - You know what happened, when, and where
        - You can save logs
        - You can control output

```python
import logging

logging.basicConfig(level=logging.DEBUG)

logging.debug("Debug message")
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
```

- Log Levels

| Level    | Purpose                       |
| -------- | ----------------------------- |
| DEBUG    | Detailed info (developer use) |
| INFO     | Normal events                 |
| WARNING  | Something unexpected          |
| ERROR    | Failure                       |
| CRITICAL | Serious failure               |


- Configuring Logging
     - This controls:
        - Where logs go (file/console)
        - What level to record
        - How logs look

```python
logging.basicConfig(
    filename='app.log',
    filemode = 'a',
    level=logging.DEBUG,
    format='%(asctime)s - %(levelname)s - %(message)s'
)
```

```python
## Using a Logger Object
# Why?
    # Helps in large projects
    # Each file/module can have its own logger

logger = logging.getLogger(__name__)
logger.info("Program started")
```

- Multiple Loggers
    - Each logger behaves differently
    - Useful in big applications

```python
logger1 = logging.getLogger("module1")
logger2 = logging.getLogger("module2")

logger1.setLevel(logging.DEBUG)
logger2.setLevel(logging.WARNING)
```

```python
## Hnadlers : its decide where logs go

logging.basicConfig(
    handlers=[
        logging.FileHandler("temp.log"),
        logging.StreamHandler()
    ]
)
```

```python
## Formatters : control how log look
# eg. : 2026-04-18 16:45:00 - app - ERROR - Something failed
format='%(asctime)s - %(name)s - %(levelname)s - %(message)s'
```

Error Handling with Logging

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    logging.error("Division by zero")
```

- Advanced Logging Concept

```python
# 1. Multiple Handlers (Different outputs)

file_handler = logging.FileHandler("app.log")
console_handler = logging.StreamHandler()

logger.addHandler(file_handler)    #add handler
logger.addHandler(console_handler)

# File → everything
# Console → only errors
```

```python
# 2. Different Levels for Handlers

file_handler.setLevel(logging.DEBUG)
console_handler.setLevel(logging.ERROR)
```

```python
# 3. Logging Exceptions
try:
    x = 1 / 0
except Exception:
    logging.exception("Exception occurred")
```

Logging system has 3 main parts:

1. Logger

    - Creates log messages

2. Handler

    - Decides where logs go

3. Formatter

    - Decides how logs look
