Multi-Threading and Multi-Processing

1. Program

A program is a sequence of instructions written in a programming language (e.g., Python, C++). 

Example: Google Chrome → Executable program → Browser functionality.

2. Process

- A process is an instance of a program that is being executed.
- Each process has a separate memory space:

    - Code
    - Segment
    - Data
    - Segment
    - Heap
    - Stack
    - Registers


- Key property: One process cannot corrupt another process.

- Examples:
    - Browser process
    - Kernel assigns process IDs
    - Applications run as processes

- Processes involve I/O requirements and resource execution.


3. Threads

- A thread is a unit of execution within a process.

- Threads share the same memory segments (Code, Data, Heap) but have separate stacks and registers.

- Single-threaded process:

    - One stack and one set of registers.

- Multi-threaded process:
    - Multiple stacks and registers for each thread.

- Example: MS Paint → Process (Box) → Threads (Circles).


4. Examples of Process and Thread

Process Example: Browser running as a separate process.

Thread Example: Multiple tabs in a browser running as threads within the same process.

Key Takeaways

Program: Static code.

Process: Running instance of a program with isolated memory.

Thread: Lightweight execution unit within a process, sharing memory but maintaining its own stack and registers.

Multi-threading allows parallel execution within the same process, while multi-processing involves independent processes with separate memory spaces.
