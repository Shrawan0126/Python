# Python Control Flow

Control flow decides **which** code runs and **how many times** it runs.

Three big buckets:
1. Conditional statements (`if` / `elif` / `else`)
2. Loops (`for`, `while`)
3. Loop control (`break`, `continue`, `pass`, `else` on loops)

Everything below has runnable cells. Read the comments, then run.


## 1. Conditional Statements

Python uses **indentation** (not braces) to define blocks. Standard = 4 spaces.

Syntax:
```python
if condition:
    # runs if condition is True
elif other_condition:
    # runs if first was False and this is True
else:
    # runs if all above were False
```
- Only the **first** matching block runs.
- `elif` and `else` are optional.


````python
# Basic if / elif / else
marks = 82

if marks >= 90:
    grade = "A+"
elif marks >= 75:        # checked only if marks < 90
    grade = "A"
elif marks >= 60:
    grade = "B"
else:
    grade = "Fail"

print("Grade:", grade)
````

### Truthy and Falsy values

Conditions don't need to be `True`/`False`. Python evaluates the **truthiness** of any object.

**Falsy:** `0`, `0.0`, `""`, `[]`, `{}`, `()`, `set()`, `None`, `False`
**Truthy:** everything else (non-empty containers, non-zero numbers, etc.)


````python
# Truthy / Falsy in action
items = []

if items:                       # empty list is Falsy
    print("List has items")
else:
    print("List is empty")      # this runs

name = "Dusker"
if name:                        # non-empty string is Truthy
    print("Hello", name)
````

### Ternary (conditional expression)

One-line if/else. Good for simple value assignment.

```python
value = A if condition else B
```


````python
# Ternary operator
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)

# Nested ternary (use sparingly - hurts readability)
n = 0
sign = "positive" if n > 0 else "negative" if n < 0 else "zero"
print(sign)
````

### `match` statement (Python 3.10+)

Structural pattern matching. Cleaner than long `elif` chains for fixed cases.
`case _` is the default (wildcard).


````python
# match-case
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 301 | 302:          # multiple patterns with |
            return "Redirect"
        case 404:
            return "Not Found"
        case _:                  # default / wildcard
            return "Unknown"

print(http_status(200))
print(http_status(302))
print(http_status(500))
````

## 2. Loops

### 2.1 `for` loop

Iterates over any **iterable** (list, string, tuple, dict, range, etc.).
Python's `for` is a *for-each*, not a counter loop.


````python
# for over a list
fruits = ["apple", "banana", "cherry"]
for f in fruits:
    print(f)

# for over a string -> iterates characters
for ch in "abc":
    print(ch)
````

### `range()` — generating number sequences

- `range(stop)` -> 0 .. stop-1
- `range(start, stop)` -> start .. stop-1
- `range(start, stop, step)` -> step can be negative

`range` is **lazy** (doesn't build the full list in memory).


````python
# range variations
print(list(range(5)))          # 0 1 2 3 4
print(list(range(2, 8)))       # 2 3 4 5 6 7
print(list(range(0, 10, 2)))   # 0 2 4 6 8
print(list(range(5, 0, -1)))   # 5 4 3 2 1  (countdown)
````

### `enumerate()` — index + value together

Cleaner than manual counters. Optional `start` argument.


````python
# enumerate gives (index, value)
colors = ["red", "green", "blue"]
for i, c in enumerate(colors):
    print(i, c)

# start counting from 1
for i, c in enumerate(colors, start=1):
    print(i, c)
````

### `zip()` — loop over multiple iterables in parallel

Stops at the **shortest** iterable.


````python
# zip pairs items by position
names = ["Yash", "Shivani", "Nitya"]
marks = [88, 92, 79]

for n, m in zip(names, marks):
    print(n, "->", m)
````

### 2.2 `while` loop

Repeats **as long as** the condition stays True.
Risk: forgetting to update the condition -> infinite loop.


````python
# while loop - must update the variable
count = 0
while count < 5:
    print("count =", count)
    count += 1          # without this -> infinite loop
````

````python
# while for input-style validation (simulated)
balance = 100
withdrawals = [30, 40, 50]   # last one fails
i = 0
while i < len(withdrawals):
    amt = withdrawals[i]
    if amt <= balance:
        balance -= amt
        print("Withdrew", amt, "| balance:", balance)
    else:
        print("Insufficient funds for", amt)
    i += 1
````

## 3. Loop Control Statements

| Keyword    | Effect                                              |
|------------|-----------------------------------------------------|
| `break`    | Exit the loop immediately                           |
| `continue` | Skip rest of current iteration, go to next          |
| `pass`     | Do nothing (placeholder)                            |
| `else`     | Runs if loop finished **without** hitting `break`   |


````python
# break - stop when target found
nums = [3, 7, 1, 9, 4]
for n in nums:
    if n == 9:
        print("Found 9, stopping")
        break
    print("checking", n)
````

````python
# continue - skip even numbers
for n in range(1, 8):
    if n % 2 == 0:
        continue        # skip the print below
    print("odd:", n)
````

````python
# pass - syntactically needed, does nothing
for n in range(3):
    pass   # placeholder, e.g. 'TODO later'
print("loop ran, body was empty")
````

### `for ... else` and `while ... else`

The `else` block runs **only if the loop completed normally** (no `break`).
Classic use: searching — `else` means "not found".


````python
# for-else: detect if item was NOT found
data = [2, 4, 6, 8]
target = 5

for x in data:
    if x == target:
        print("Found")
        break
else:
    print("Not found")   # runs because break never happened
````

## 4. Nested Loops

A loop inside a loop. Inner loop runs fully for **each** outer iteration.
Common for grids, matrices, pattern printing.


````python
# Nested loop - multiplication-style grid
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i}x{j}={i*j}", end="  ")
    print()   # newline after each row
````

## 5. Comprehensions (Pythonic looping)

Compact way to build lists/dicts/sets in one line.
Faster and cleaner than equivalent loops for simple transforms.

```python
[expression for item in iterable if condition]
```


````python
# List comprehension
squares = [n*n for n in range(1, 6)]
print(squares)

# With condition (filter)
evens = [n for n in range(10) if n % 2 == 0]
print(evens)

# Dict comprehension
sq_map = {n: n*n for n in range(1, 5)}
print(sq_map)

# Set comprehension (auto-removes duplicates)
unique = {ch for ch in "mississippi"}
print(unique)
````

## 6. Quick Recap / Cheat Sheet

- `if / elif / else` -> branch on conditions; only first match runs
- Truthy/Falsy -> empty + zero + None are False
- Ternary -> `A if cond else B`
- `match` -> clean multi-case (3.10+)
- `for` -> iterate any iterable; pair with `range`, `enumerate`, `zip`
- `while` -> condition-based; remember to update it
- `break` / `continue` / `pass` -> control flow inside loops
- `for...else` -> else runs if no break (great for search)
- Comprehensions -> one-line list/dict/set building

**Common mistakes**
- Forgetting `:` after `if`/`for`/`while`
- Wrong indentation
- Infinite `while` (no update)
- Using `=` (assign) instead of `==` (compare)
