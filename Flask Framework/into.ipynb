# Flask Framework – Architecture Notes

## Overview
Flask is a lightweight Python web framework used to build web applications and APIs.  
It acts as the application layer that processes client requests and returns responses.

---

## Request–Response Flow

1. Client sends multiple requests (Req1, Req2, Req3)  
2. Requests are received by the Web Server  
3. Web Server forwards requests to the Python application using WSGI  
4. Flask Web Application processes the request  
5. Based on the route (/home, /api, /functionalities), a response is generated  
6. Response is sent back to the client via the Web Server  

---

## Components

### Web Server
- Handles incoming HTTP requests  
- Examples: Apache, IIS  
- Can be deployed on cloud platforms  

### Hosting / Deployment Platforms
- AWS EC2  
- Azure App Services  
- Apache  
- IIS  

### WSGI (Web Server Gateway Interface)
- Interface between Web Server and Python application  
- Acts as a bridge for communication  
- Ensures standardized request handling  

### Flask Web Application
- Contains application logic  
- Handles routing and functionality  
- Example routes:  
  - /home  
  - /api  
  - /functionalities  

---

## Jinja2 Template Engine

### Purpose
- Used by Flask to render dynamic web pages  
- Embeds Python-like expressions inside HTML templates  

### Data Sources
- SQL Databases  
- CSV Files  
- Machine Learning Models  
- MongoDB  

### Flow
1. Web template defines the structure/layout of the page  
2. Data source provides dynamic content  
3. Jinja2 combines template and data to generate the final HTML page  

### Example Use Cases
- Login pages with database integration  
- Displaying ML model predictions  
- Dashboards with dynamic data  

---

## Key Points
- Web Server does not run Python code directly  
- WSGI connects Web Server and Flask  
- Flask handles logic and generates responses  
- Jinja2 enables dynamic rendering of content from various data sources  

---

## Summary
Flow:  
Client → Web Server → WSGI → Flask App → Jinja2 → Response
