# Python Inbuilt Data Structures — Advanced Notes

Python ships with 4 core built-in data structures:

| Structure | Ordered | Mutable | Duplicates | Syntax       |
|-----------|---------|---------|------------|--------------|
| List      | Yes     | Yes     | Allowed    | `[ ]`        |
| Tuple     | Yes     | No      | Allowed    | `( )`        |
| Set       | No*     | Yes     | Not allowed| `{ }`        |
| Dict      | Yes**   | Yes     | Keys unique| `{key: val}` |

\* sets have no index order  
\** dicts keep insertion order (Python 3.7+)

All cells are runnable. Read comments, then run.


## 1. List

Ordered, mutable, allows duplicates, can mix types.
The workhorse of Python.


````python
# Creating lists
empty = []
nums = [1, 2, 3, 4, 5]
mixed = [1, "two", 3.0, True]      # types can mix
nested = [[1, 2], [3, 4]]          # list of lists
print(nums, mixed, nested)
````

### Indexing & Slicing

- Index starts at `0`; negative index counts from the end (`-1` = last).
- Slice: `list[start:stop:step]` -> stop is **excluded**.


````python
nums = [10, 20, 30, 40, 50]

print(nums[0])      # 10  (first)
print(nums[-1])     # 50  (last)
print(nums[1:4])    # [20, 30, 40]  (index 1..3)
print(nums[:3])     # [10, 20, 30]  (start omitted)
print(nums[::2])    # [10, 30, 50]  (every 2nd)
print(nums[::-1])   # reversed list
````

### Common List Methods

| Method            | What it does                         |
|-------------------|--------------------------------------|
| `append(x)`       | add x at end                         |
| `insert(i, x)`    | insert x at index i                  |
| `extend(iter)`    | add all items from iterable          |
| `remove(x)`       | remove first occurrence of x         |
| `pop(i)`          | remove & return item at i (last if no i) |
| `index(x)`        | first index of x                     |
| `count(x)`        | how many times x appears             |
| `sort()`          | sort in place                        |
| `reverse()`       | reverse in place                     |
| `copy()`          | shallow copy                         |
| `clear()`         | empty the list                       |


````python
lst = [3, 1, 2]

lst.append(4)            # [3, 1, 2, 4]
lst.insert(0, 0)         # [0, 3, 1, 2, 4]
lst.extend([5, 6])       # [0, 3, 1, 2, 4, 5, 6]
print("after adds:", lst)

lst.remove(0)            # removes value 0
last = lst.pop()         # removes & returns 6
print("popped:", last, "| now:", lst)

lst.sort()               # ascending
print("sorted:", lst)
lst.sort(reverse=True)   # descending
print("desc:", lst)
````

### Mutability & the aliasing trap

Lists are **mutable** and assignment copies the *reference*, not the data.
Use `.copy()` or `list(x)` for an independent copy.


````python
a = [1, 2, 3]
b = a            # b points to SAME list
b.append(4)
print("a:", a)   # a also changed! [1,2,3,4]

c = a.copy()     # independent copy
c.append(99)
print("a:", a)   # unchanged
print("c:", c)
````

## 2. Tuple

Ordered, **immutable**, allows duplicates.
Use when data should not change (coordinates, fixed records, dict keys).
Slightly faster + memory-light vs lists.


````python
# Creating tuples
t = (1, 2, 3)
single = (5,)            # NOTE: comma needed for 1-element tuple
no_paren = 1, 2, 3       # parentheses optional
print(t, single, type(no_paren))
````

````python
# Tuples are immutable
t = (10, 20, 30)
print(t[0], t[-1])       # indexing works

try:
    t[0] = 99            # this fails
except TypeError as e:
    print("Error:", e)
````

### Tuple unpacking

Assign multiple variables at once. `*` captures the rest.


````python
# Unpacking
point = (3, 7)
x, y = point
print("x:", x, "y:", y)

# Star unpacking
a, *middle, z = (1, 2, 3, 4, 5)
print(a, middle, z)      # 1 [2,3,4] 5

# Swap variables (classic Python trick)
p, q = 1, 2
p, q = q, p
print("swapped:", p, q)
````

### Tuple methods (only two)

`count(x)` and `index(x)` — because tuples can't be modified.


````python
t = (1, 2, 2, 3, 2)
print("count of 2:", t.count(2))
print("index of 3:", t.index(3))
````

## 3. Set

Unordered, mutable, **no duplicates**.
Great for: removing duplicates, membership tests (very fast), set math.
NOTE: `{}` creates an empty **dict**, not a set — use `set()`.


````python
# Creating sets
s = {1, 2, 3, 3, 2}      # duplicates auto-removed
print(s)                 # {1, 2, 3}

empty = set()            # NOT {}
print(type(empty))

# Quick dedupe of a list
nums = [1, 1, 2, 3, 3, 3]
print(list(set(nums)))
````

### Set methods

| Method        | Effect                          |
|---------------|---------------------------------|
| `add(x)`      | add one element                 |
| `update(it)`  | add many elements               |
| `remove(x)`   | remove x (error if missing)     |
| `discard(x)`  | remove x (no error if missing)  |
| `pop()`       | remove arbitrary element        |


````python
s = {1, 2, 3}
s.add(4)               # {1,2,3,4}
s.update([5, 6])       # {1,2,3,4,5,6}
s.discard(10)          # no error even though 10 absent
s.remove(1)            # removes 1
print(s)
````

### Set operations (math)

| Operation     | Operator | Method                  |
|---------------|----------|-------------------------|
| Union         | `\|`     | `a.union(b)`            |
| Intersection  | `&`      | `a.intersection(b)`     |
| Difference    | `-`      | `a.difference(b)`       |
| Symmetric diff| `^`      | `a.symmetric_difference(b)` |


````python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print("union:", a | b)            # all elements
print("intersection:", a & b)     # common
print("difference:", a - b)       # in a, not in b
print("symmetric:", a ^ b)        # in either, not both

# Membership test - very fast (O(1) average)
print(3 in a)
````

## 4. Dictionary

Key-value pairs. Keys must be **unique** and **hashable** (immutable types).
Insertion order preserved (3.7+). Lookups are very fast.


````python
# Creating dicts
student = {"name": "Shivani", "age": 20, "branch": "CSE"}
empty = {}
from_pairs = dict([("a", 1), ("b", 2)])
print(student)
print(from_pairs)
````

### Accessing & updating

- `d[key]` -> raises `KeyError` if missing
- `d.get(key, default)` -> safe, returns default if missing


````python
student = {"name": "Shivani", "age": 20}

print(student["name"])            # direct access
print(student.get("age"))         # safe
print(student.get("phone", "N/A"))# missing -> default

student["age"] = 21               # update existing
student["city"] = "Ajmer"         # add new key
print(student)
````

### Dict methods

| Method            | Returns / Effect                       |
|-------------------|----------------------------------------|
| `keys()`          | view of keys                           |
| `values()`        | view of values                         |
| `items()`         | view of (key, value) pairs             |
| `get(k, default)` | safe lookup                            |
| `pop(k)`          | remove key, return its value           |
| `update(d2)`      | merge another dict                     |
| `setdefault(k,v)` | get k, or set it to v if absent        |


````python
d = {"a": 1, "b": 2, "c": 3}

print(list(d.keys()))
print(list(d.values()))
print(list(d.items()))

d.update({"d": 4, "a": 99})      # merge + overwrite
print("after update:", d)

val = d.pop("b")                 # remove 'b'
print("popped:", val, "| now:", d)
````

### Looping over a dict

By default iterating a dict gives **keys**. Use `.items()` for both.


````python
scores = {"Yash": 88, "Nitya": 92, "Pooja": 79}

# keys only
for name in scores:
    print(name)

print("---")

# key + value
for name, score in scores.items():
    print(name, "->", score)
````

### Nested dictionaries

Dicts inside dicts — common for JSON-like data.


````python
team = {
    "opt1": {"name": "Yash", "marks": 88},
    "opt2": {"name": "Nitya", "marks": 92},
}

print(team["opt2"]["name"])      # access nested value

for key, info in team.items():
    print(key, ":", info["name"], info["marks"])
````

## 5. When to Use What

- **List** -> ordered collection you'll modify; sequences, stacks, queues
- **Tuple** -> fixed data that shouldn't change; safe to use as dict keys
- **Set** -> uniqueness + fast membership + set math (union/intersection)
- **Dict** -> labeled data, fast key lookup, JSON-like structures

### Mutability summary
- Mutable: `list`, `set`, `dict`
- Immutable: `tuple` (also `str`, `int`, `float`, `frozenset`)

### Performance notes
- Membership `x in ...`: **set/dict O(1)** avg, **list O(n)**
- Use a set/dict when you do many lookups


## 6. Cheat Sheet

```python
# List
lst = [1, 2, 3]; lst.append(4); lst[0]; lst[::-1]

# Tuple
t = (1, 2, 3); a, b, c = t            # immutable + unpack

# Set
s = {1, 2, 3}; s.add(4); a & b; a | b # unique + math

# Dict
d = {"k": "v"}; d.get("k"); d.items() # key-value lookup
```

**Common mistakes**
- `{}` makes an empty dict, not a set -> use `set()`
- 1-element tuple needs a comma: `(5,)` not `(5)`
- `b = a` shares the same list -> use `a.copy()`
- `d[missing_key]` raises KeyError -> use `.get()`
