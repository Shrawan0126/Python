```python
try:
    result = 1/0
except ZeroDivisionError as ex:
    print(ex)
    print("Cannot divide by zero.")
```

```python
### Handling multiple exceptions in a single block
try:
    result = 1/2
    a = b
except ZeroDivisionError as ex:
    print(ex)
    print("Cannot divide by zero.")
except NameError as e:
    print(f"An error occurred: {e}")
```

```python
### Handling multiple exceptions in a single block
try:
    number = int(input("Enter a number: "))
    result = 10 / number
except ValueError as ex:
    print(f"Invalid input: {ex}")
except ZeroDivisionError as ex:
    print(f"Cannot divide by zero: {ex}")
except Exception as ex:
    print(f"An unexpected error occurred: {ex}")
```

```python
### try, except, else block
try:
    number = int(input("Enter a number: "))
    result = 10 / number
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")
except Exception as ex:
    print(f"An unexpected error occurred: {ex}")    
else:
    print(f"The result is: {result}")
```

```python
### try, except, else, and finally
try:
    number = int(input("Enter a number: "))
    result = 10 / number  
except ValueError:
    print("Invalid input. Please enter a valid number.")
except ZeroDivisionError:
    print("Cannot divide by zero.")  
except Exception as ex:
    print(f"An unexpected error occurred: {ex}")
else:                                           ## This block will execute only if no exceptions were raised in the try block.
    print(f"The result is: {result}")          
finally:
    print("This block will always execute, regardless of exceptions.")  
```

```python
### File handling with exceptions
try:  
    file = open("non_existent_file.txt", "r")
    content = file.read()
    a = b
except FileNotFoundError as ex:
    print("This file does not exist.")
except Exception as ex:
    print(f"An error occurred: {ex}")
finally:
    if 'file' in locals() and not file.closed:
        file.close()
        print("File has been closed.")
```
