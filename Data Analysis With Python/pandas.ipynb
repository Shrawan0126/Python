pandas is a powerful data manipilation library in Python, widely used for data analysis and data cleaning. It provides two primary data structures: Series and DataFrame. A Series is a one - dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable.
Use:
Working with data (tables, Excel, CSV)
Data analysis and cleaning
Handling rows and columns

```python
import pandas as pd
```

```python
## Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.
data = [1, 2, 3, 4, 5]
s = pd.Series(data)
print(s)
```

```python
## Create a series from Directionary
data = {'a': 1, 'b': 2, 'c': 3}
s = pd.Series(data)  
print(s)
```

```python
data = [10,20,30]
index = ['a', 'b', 'c'] 
s = pd.Series(data, index=index)
print(s)
```

```python
## DataFrame : A DataFrame is a two-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet or SQL table, and it is one of the most commonly used data structures in pandas.
## Create a Dataframe from a dictionary of list
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'Los Angeles', 'Chicago']
        }
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
## Create a Data frame From a list of Dictionaries
data = [{'Name': 'Alice', 'Age': 25, 'City': 'New York'},
        {'Name': 'Bob', 'Age': 30, 'City': 'Los Angeles'},
        {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'}
        ]
df = pd.DataFrame(data)
print(df)
```
