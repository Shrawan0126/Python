**Data Structures in Pandas**

Pandas provides two data structures for manipulating data which are as follows:

**Pandas Series**
- One-dimensional labeled array
- Holds data of a single type (int, float, string, etc.)
- Each element has an index (label)
- Think of it as a single column in Excel
- Example: Stock prices of one company

**Pandas DataFrame**
- Two-dimensional labeled data structure
- Consists of multiple Series aligned by rows
- Can hold heterogeneous data (different types in different columns)
- Think of it as a full spreadsheet with rows and columns
- Example: Dataset with City, State, Population, Income

**Key Differences**
- Dimensionality: Series → 1D, DataFrame → 2D
- Composition: DataFrame is a collection of Series
- Use Case: Series for single variable, DataFrame for multiple variables

### **Operations in Pandas**

**1. Loading Data**: This operation reads data from files such as CSV, Excel or JSON into a DataFrame.

```python
import pandas as pd

df = pd.read_csv("data.csv")
print(df.head())
```

**2. Viewing and Exploring Data**: After loading data, it is important to understand its structure and content. This methods allow you to inspect rows, summary statistics and metadata.

```python
print(df.info())  # To get the information about the DataFrame
print(df.describe())  # To get the statistical summary of the DataFrame
```

**3. Handling Missing Data:** Datasets often contain empty or missing values. Pandas provides functions to detect, remove or replace these values.

```python
print(df.isnull().sum())
df = df.fillna(0)     # Fill missing values with 0
```

**4. Selecting and Filtering Data:** This operation retrieves specific columns, rows or records that match a condition. It allows precise extraction of required information.

```python
import pandas as pd

df = pd.read_csv("data.csv", quotechar="'")

# Strip single quotes from column names
df.columns = df.columns.str.strip("'")

# Convert age column to integers
df['age'] = df['age'].astype(int)

# Now filtering works
ages = df[df['age'] > 25]
print(ages)
```

**5. Adding and Removing Columns:** You can create new columns based on existing ones or delete unwanted columns from the DataFrame.

```python
#Add a new column
df['total'] = df['a'] + df['b']
print(df.head())

print("\n")

#Remove a column
df = df.drop('total', axis=1)
print(df.head())
```

**6. Grouping Data (GroupBy):** Grouping allows you to organize data into categories and compute values for each group for example, sums, counts or averages.

```python
res = df.groupby('category')['sales'].sum()
print(res)
```

#### Creating DataFrame from dict of Numpy Array

```python
import numpy as np
import pandas as pd

data = { 'A': np.array([1, 4, 7]),
         'B': np.array([2, 5, 8]),
         'C': np.array([3, 6, 9]) }
df = pd.DataFrame(data)
print(df)
```

#### Creating a DataFrame from a List of Dictionaries

```python
import pandas as pd
data = [
    {'name': 'Mike', 'degree': 'MBA', 'score': 90},
    {'name': 'Dan', 'degree': 'BCA', 'score': 40},
    {'name': 'Emilia', 'degree': 'M.Tech', 'score': 80},
]
df = pd.DataFrame(data)
print(df)
```
