Reading Data From Different Sources

```python
import pandas as pd
from io import StringIO

Data = '[{"employee_name": "James", "employee_id": 1, "employee_salary": 50000, "employee_age": 30, "profile_image": ""}]'
df = pd.read_json(StringIO(Data))

print(df)
```

```python
# Convert DataFrame to JSON format with 'records' orientation
df.to_json(orient='records')
```

```python
df.to_json(orient='index')
```

```python
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header=None)
df.head()
```

```python
#convert into .csv file
df.to_csv("wine.csv")
```

```python
df = pd.read_html("https://statisticstimes.com/sports/ipl/all-ipl-points-table.php",header=0)
df
```

```python
url = "https://www.fdic.gov/bank/individual/failed/banklist.html"
df = pd.read_html(url)
```
