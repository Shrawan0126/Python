```python
import numpy as np

# np.dtype creates a data type object in NumPy.
# This object tells NumPy what kind of values an array will store.

print(np.dtype(np.int16))   # Using np.int16 (16-bit signed integer type)
                            # Range: -32768 to 32767
                            # Each value takes 2 bytes in memory

print(np.dtype('int16'))    # Same as above, but dtype specified as a string
                            # Both methods return the same dtype: int16
```

```python
import numpy as np

# Create a dtype object:
# '>i4' means:
#   '>'  → Big-endian byte order (most significant byte stored first)
#   'i'  → Integer type
#   '4'  → 4 bytes (32-bit integer)

dt = np.dtype('>i4')

# dt.byteorder shows the byte order:
#   '>' means big-endian
print("Byte order:", dt.byteorder)

# dt.itemsize shows how many bytes each item takes:
#   4 bytes = 32 bits
print("Size:", dt.itemsize)

# dt.name gives the human-readable name of the dtype:
#   'int32' because 4 bytes = 32-bit integer
print("Data type:", dt.name)
```

```python
import numpy as np

# Define a structured dtype with two fields:
# 'name'   → Unicode string of length 16 (written as 'U16')
# 'scores' → Array of 2 float64 values
dt = np.dtype([
    ('name', 'U16'),              # Unicode string, max length 16
    ('scores', np.float64, (2,))  # 2-element array of float64
])

# Access dtype information for each field
print(dt['name'])    # <U16 → Unicode string of length 16
print(dt['scores'])  # (2,)float64 → array of 2 float64 values
```

#### Creating a structured array

```python
import numpy as np

# Define a structured dtype:
# 'name'   → Unicode string of length 16
# 'scores' → Array of 2 float64 values
dt = np.dtype([
    ('name', 'U16'),              # Correct way to specify Unicode string length
    ('scores', np.float64, (2,))  # 2-element array of float64
])

# Create an array with this dtype
data = np.array([
    ('Emma', (8.5, 7.0)),
    ('Lucas', (6.0, 7.5))
], dtype=dt)

# Access elements
print(data[1])              # Prints the full record for 'Lucas'
print("Scores:", data[1]['scores'])  # Prints Lucas's scores → [6.0 7.5]
print("Names:", data['name'])        # Prints all names → ['Emma' 'Lucas']
```
