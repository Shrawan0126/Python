```python
## Handling Missing Values
# Check for missing values in the dataset
df.isnull().any(axis=1)
```

```python
# Count the number of missing values in each column
df.isnull().sum()
```

```python
### Filling missing values with the mean of the column
df['Total Revenue'].fillna(df['Total Revenue'].mean(), inplace=True)
```

```python
## Renaming Columns
df = df.rename(columns={'Total Revenue': 'Revenue', 'Units Sold': 'Units'})
df.head()
```

```python
df['New Value'] = df['Revenue'].apply(lambda x : x*2)
df.head()
```

```python
grouped_mean = df.groupby('Product Category')['Revenue'].mean()
print(grouped_mean)
```

```python
## Aggregate multiple nulriple funnctions
grouped_agg = df.groupby('Region')['Revenue'].agg(['mean', 'sum'])
grouped_agg
```

```python
### Merging and joining Dataframes
df1 = pd.DataFrame({'key' : ['A', 'B', 'C'], 'value1' : [1, 2, 3]})
df2 = pd.DataFrame({'key' : ['A', 'B', 'D'], 'value2' : [4, 5, 6]})
```

```python
df1
```

```python
df2
```

```python
## Merge the dataframe on the 'key' column
pd.merge(df1,df2, on = 'key', how = 'inner')
```

```python
pd.merge(df1,df2, on = 'key', how = 'outer')
```

```python
pd.merge(df1,df2,on="key",how="left") #priority to df1
```

```python
pd.merge(df1,df2,on="key",how="right") #priority to df2
```
