```python
### Read a whole File

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
### Read a file line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line) 
```

```python
# strip() removes leading/trailing whitespace
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

```python
### Writing a file(Overwriting)
with open('example.txt', 'w') as file:
    file.write("This is a new file.\n")
    file.write("It contains some text.")
```

```python
### Write a file(without overwriting)
with open('example.txt', 'a') as file:
    file.write("\nThis line is appended to the file.")
```

```python
### Writing a list of lines to a file
lines = ["\nFirst line.\n", "Second line.\n", "Third line.\n"]
with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Read from source and write to destination
with open('example.txt', 'r') as src_file:
    content = src_file.read()
with open('destination.txt', 'w') as dest_file:
        dest_file.write(content)
```

```python
### Read a text file and count lines, words, and characters
def count_file_contents(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        num_lines = len(lines)
        num_words = sum(len(line.split()) for line in lines)
        num_characters = sum(len(line) for line in lines)
    return num_lines, num_words, num_characters

path = 'example.txt'
lines,words,characters = count_file_contents(path)
print(f"Lines: {lines}, Words: {words}, Characters: {characters}")
```

```python
### Writing and then reading a file
with open('example.txt', 'w+') as file:  # 'w+' allows both writing and reading
    file.write("Hello, World!\nThis is a test file.")
    file.seek(0)  # Move the cursor back to the beginning of the file
    content = file.read()
    print(content)
```
