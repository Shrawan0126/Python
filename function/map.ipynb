```python
def square(x):
    return x * x
square(10)
```

```python
numbers = [1, 2, 3, 4, 5]
list(map(square, numbers))
```

```python
## Lambda function with map
numbers = [1, 2, 3, 4, 5]
list(map(lambda x: x * x, numbers))
```

```python
## map multiple iterables
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]
added_numbers = list(map(lambda x, y: x + y, numbers1, numbers2))
print(added_numbers)
```

```python
## convert a list of strings to inegers
# use map to convert a list of strings to integers
string_numbers = ['1', '2', '3', '4', '5']
integer_numbers = list(map(int, string_numbers))
print(integer_numbers) 
```

```python
## use map to convert a list of strings to uppercase
words = ['hello', 'world', 'python']
uppercase_words = list(map(str.upper, words))
print(uppercase_words)
```

```python
def get_name(person):
    return person['name']
people = [{'name': 'Alice', 'age': 30},
          {'name': 'Bob', 'age': 25},
          {'name': 'Charlie', 'age': 35}]
list(map(get_name, people))
```
